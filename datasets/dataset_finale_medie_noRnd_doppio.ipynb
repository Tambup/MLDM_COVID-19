## Creazione dataset denominato: dataset_finale_medie.csv

In questo dataset le X sono composte da tutti i parametri presi sei giorni stocasticamente compresi fra le due settimane precedenti. Di questi dati viene computata la media aritmetica.

```
col=list(dataset_completo.columns)
col.remove("Data")
even_indexes=np.array([2,4,6,8,10,12])
odd_indexes=np.array([3,5,7,9,11,13])

for i in range(0,dataset_completo.shape[0]-18):
  data_previsione=dataset_completo.loc[i+14, "Data"]
  even_media_passato=dataset_completo.iloc[even_indexes+i].mean().to_frame()
  odd_media_passato=dataset_completo.iloc[odd_indexes+i].mean().to_frame()
  
  even_media_passato=even_media_passato.pivot_table(columns=col).add_prefix("pass_")
  odd_media_passato=odd_media_passato.pivot_table(columns=col).add_prefix("pass_")
  even_media_passato["Data"]=data_previsione
  odd_media_passato["Data"]=data_previsione
  fut1=pd.to_numeric(dataset_completo.loc[i+16, dataset_completo.columns != "Data"]).to_frame().pivot_table(columns=col).add_prefix("fut1_").reset_index()
  fut2=pd.to_numeric(dataset_completo.loc[i+18, dataset_completo.columns != "Data"]).to_frame().pivot_table(columns=col).add_prefix("fut2_").reset_index()
  even_riga=pd.concat([even_media_passato, fut1, fut2], axis=1).drop(columns="index")
  odd_riga=pd.concat([odd_media_passato, fut1, fut2], axis=1).drop(columns="index")
  if i==0:
    finale=pd.DataFrame(columns=even_riga.columns)
  
  finale=pd.concat([finale, even_riga], axis=0)
  finale=pd.concat([finale, odd_riga], axis=0)
```

Qui per comodità si sposta "Data" come primo parametro. 

```
colonne_non_ordinate=list(finale.columns)
colonne_non_ordinate.remove("Data")
elenco_ordinato_colonne=["Data"]

elenco_ordinato_colonne=elenco_ordinato_colonne+colonne_non_ordinate
finale=finale.reindex(columns=elenco_ordinato_colonne)
finale
```

```
finale.to_csv("dataset_finale_medie_doppio.csv", index=False)
```
